## 1Ô∏è‚É£ Imports & Chargement des donn√©es
On charge tous les datasets g√©n√©r√©s par le pipeline d'acquisition.

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import numpy as np

import fastf1
```

```python
DATA_DIR = '../data'

df_results = pd.read_parquet(f'{DATA_DIR}/results_2025.parquet')
df_quali = pd.read_parquet(f'{DATA_DIR}/qualifying_2025.parquet')
df_pits = pd.read_parquet(f'{DATA_DIR}/pitstops_2025.parquet')
df_drv_stand = pd.read_parquet(f'{DATA_DIR}/driver_standings_2025.parquet')
df_team_stand = pd.read_parquet(f'{DATA_DIR}/team_standings_2025.parquet')
df_weather = pd.read_parquet(f'{DATA_DIR}/weather_2025.parquet')
df_flights = pd.read_parquet(f'{DATA_DIR}/flightlegs_2025.parquet')
```

## 2Ô∏è‚É£ Statistiques descriptives & premi√®res analyses
Aper√ßu rapide des principales tables : stats, shape, types, valeurs manquantes.

```python
# Statistiques descriptives de base sur les principaux DataFrames

print("----- R√©sultats course -----")
display(df_results.describe())

print("\n----- Qualifications -----")
display(df_quali.describe())

print("\n----- Pitstops -----")
display(df_pits.describe(include='all'))

print("\n----- Classement pilotes -----")
display(df_drv_stand.describe(include='all'))

print("\n----- Classement √©quipes -----")
display(df_team_stand.describe(include='all'))
```

## 3Ô∏è‚É£ Corr√©lations & KPIs cl√©s
On explore les relations entre variables, la strat√©gie pneus, les abandons et l'empreinte carbone.

```python
# Quelques relations/corr√©lations int√©ressantes √† explorer
 
print("\nCorr√©lation GridPosition / Position (r√©sultats course)")
display(df_results[['GridPosition', 'Position']].corr())

print("\nPoints cumul√©s par √©curie :")
display(df_results.groupby('TeamName')['Points'].sum().sort_values(ascending=False))

print("\nR√©partition des compounds utilis√©s aux stands :")
if 'CompoundIn' in df_pits.columns:
    display(df_pits['CompoundIn'].value_counts())
else:
    print("CompoundIn non trouv√© dans df_pits.")

# Nombre d'abandons (Status != 'Finished' ou √©quivalent)
if 'Status' in df_results.columns:
    print("\nNombre d'abandons/disqualifications par course :")
    abandons = df_results[~df_results['Status'].isin(['Finished', 'Lapped'])]
    display(abandons.groupby('event')['FullName'].count())
else:
    print("\nColonne Status absente du DataFrame results.")
    
print("\nTotal CO‚ÇÇ logistique F1 (en tonnes) :", df_flights['CO2_tonnes'].sum())
print("\nTop 3 plus longs trajets :")
display(df_flights.sort_values('distance_km', ascending=False).head(3))
```

## 4Ô∏è‚É£ Visualisations avanc√©es
On passe √† la data viz interactive !

### üî• Heatmap progression grille/arriv√©e par GP
Visualisation des places gagn√©es/perdues par pilote et par course.

```python
df_results['delta'] = df_results['GridPosition'] - df_results['Position']
pivot = df_results.pivot_table(
    index='FullName',
    columns='event',
    values='delta'
)

# On cr√©e les matrices pour afficher grid et arrival en hover
gridpos = df_results.pivot_table(index='FullName', columns='event', values='GridPosition')
arrpos  = df_results.pivot_table(index='FullName', columns='event', values='Position')

hovertext = np.empty(pivot.shape, dtype=object)
for i, name in enumerate(pivot.index):
    for j, ev in enumerate(pivot.columns):
        val = pivot.iloc[i, j]
        grid = gridpos.iloc[i, j]
        arr  = arrpos.iloc[i, j]
        hovertext[i, j] = (
            f"Grand Prix : <b>{ev}</b><br>"
            f"Pilote : <b>{name}</b><br>"
            f"Positions gagn√©es/perdues : <b>{'+' if pd.notna(val) and val > 0 else ''}{val if pd.notna(val) else 'N/A'}</b><br>"
            f"Position d√©part : <b>{int(grid) if pd.notna(grid) else 'N/A'}</b><br>"
            f"Position arriv√©e : <b>{int(arr) if pd.notna(arr) else 'N/A'}</b>"
        )

fig = go.Figure(
    data=go.Heatmap(
        z=pivot.values,
        x=pivot.columns,
        y=pivot.index,
        customdata=hovertext,
        hovertemplate='%{customdata}<extra></extra>',
        colorscale='Aggrnyl', 
        colorbar=dict(
            title="Places<br>gagn√©es/perdues",
            thickness=18,
            tickfont=dict(color='#F2F2F2', size=14),
            title_font=dict(color='#F2F2F2', size=16)
        ),
        reversescale=False,
        zmin=-12,  # Ajustez selon vos valeurs min/max pour un meilleur contraste
        zmax=12,
        text= np.where(np.isnan(pivot.values), "DNF/DSQ", pivot.values),
        texttemplate="%{text}"
    )
)

fig.update_layout(
    title="<b>Progression sur la grille : du d√©part √† l'arriv√©e par Grand Prix</b>",
    xaxis_title="Grand Prix",
    yaxis_title="Pilotes",
    template="plotly_dark",
    font=dict(family="Montserrat, Arial", size=15, color="#F2F2F2"),
    height=700, width=1150,
    margin=dict(l=120, r=50, t=80, b=40),
    xaxis=dict(
        showgrid=False,
        tickangle=30,
        linecolor='#444',
        tickfont=dict(size=13, color='#F2F2F2')
    ),
    yaxis=dict(
        showgrid=False,
        linecolor='#444',
        tickfont=dict(size=13, color='#F2F2F2')
    )
)
fig.show()
```

### ‚è±Ô∏è Analyse des temps au tour ‚Äì Top 4 du dernier Grand Prix

On zoome sur la performance pure !  
Ce graphique affiche **l‚Äô√©volution des temps au tour** des 4 meilleurs pilotes du dernier GP couru.  
- **Chaque courbe** = un pilote du top 4 (couleur officielle de l‚Äô√©curie)
- **Axe X** : num√©ro de tour
- **Axe Y** : temps au tour (en secondes)
- **Hover** : d√©tail pilote, √©curie, tour, chrono

*NB : Les donn√©es sont extraites dynamiquement via FastF1, donc ce visuel s‚Äôadapte √† chaque nouvelle course ajout√©e au dataset.*

```python
# 1. Dernier GP couru
last_gp_round = df_results['round'].max()
last_gp_name = df_results[df_results['round'] == last_gp_round]['event'].iloc[0]

# 2. Top 4 pilotes du dernier GP
df_last_gp = df_results[df_results['round'] == last_gp_round].sort_values('Position')
drivers = df_last_gp['Abbreviation'].head(5).tolist()
driver_to_team = df_last_gp.set_index('Abbreviation')['TeamName'].to_dict()

# 3. Palette F1 : TeamName -> couleur officielle
team_colors = df_results.drop_duplicates("TeamName").set_index("TeamName")["TeamColor"].apply(
    lambda c: f"#{c}" if not str(c).startswith("#") else str(c)).to_dict()

# 4. Charger session FastF1 (cache rapide)
session = fastf1.get_session(2025, last_gp_name, 'R')
session.load()

# 5. Cr√©er le figure Plotly
fig = go.Figure()

for driver in drivers:
    team = driver_to_team[driver]
    laps = session.laps.pick_drivers(driver).pick_quicklaps()
    if laps.empty:
        continue
    fig.add_trace(go.Scatter(
        x=laps['LapNumber'],
        y=laps['LapTime'].dt.total_seconds(),
        mode='lines+markers',
        name=driver,
        line=dict(color=team_colors.get(team, "#888"), width=3),
        marker=dict(size=6, symbol="circle"),
        hovertemplate=(
            f"<b>Pilote : {driver}</b><br>"
            f"√âcurie : {team}<br>"
            "Tour : %{x}<br>"
            "Temps : %{y:.3f} s"
        ),
    ))

fig.update_layout(
    title=f"‚è±Ô∏è Temps au tour ‚Äî Top 5 pilotes ‚Äî GP {last_gp_name} (2025)",
    xaxis_title="Num√©ro de tour",
    yaxis_title="Temps au tour (s)",
    legend_title="Pilote",
    template="plotly_dark",
    height=550,
    font=dict(family="Montserrat, Arial", size=15),
    #plot_bgcolor="#181A20",
    #paper_bgcolor="#181A20",
    margin=dict(l=60, r=40, t=60, b=40)
)

fig.show()
```

### üõû Analyse strat√©gie pneus par √©curie
Nombre d'arr√™ts par type de gomme et par √©quipe tout au long de la saison.

```python
# => Focus sur les pitstops valides uniquement
pits_valid = df_pits.dropna(subset=['CompoundIn', 'CompoundOut'])

# Par exemple : nombre d'arr√™ts par √©curie et type de gomme entr√©e
bar_data = pits_valid.groupby(['TeamName', 'CompoundIn']).size().reset_index(name='NbArrets')

# Palette officielle F1 pour les types de pneus (exemple)
compound_colors = {
    'SOFT': '#FF2B2B',         # Rouge
    'MEDIUM': '#FFD12E',       # Jaune
    'HARD': '#FFFFFF',         # Blanc
    'INTERMEDIATE': '#43B047', # Vert
    'WET': '#0067AD',          # Bleu
    'SUPER_SOFT': '#FF6F61',   # Rose
    'ULTRA_SOFT': '#A020F0',   # Violet
}

fig = px.bar(
    bar_data, x='TeamName', y='NbArrets', color='CompoundIn',
    title="Nombre d'arr√™ts par √©curie et type de pneu (entr√©e)",
    labels={'NbArrets': "Nombre d'arr√™ts", 'TeamName': '√âcurie', 'CompoundIn': 'Pneu mont√©'},
    color_discrete_map=compound_colors
)
fig.update_layout(barmode='stack', template='plotly_dark', height=600)
fig.show()
```

### üìà Bump chart ‚Äì √©volution du top 10 pilotes
Visualisation de la dynamique du championnat.

```python
# Bump chart ‚Äì √©volution du classement pilotes (top 10)
top10 = (df_drv_stand
         .groupby(['round', 'FullName'])['Points']
         .max()
         .reset_index()
         .sort_values(['round', 'Points'], ascending=[True, False])
         .groupby('round')
         .head(10))

round_to_gp = df_results.set_index('round')['event'].to_dict()
top10['event'] = top10['round'].map(round_to_gp)

# On r√©cup√®re la position finale de chaque pilote √† chaque course
position_map = df_results.set_index(['FullName', 'round'])['Position'].to_dict()
top10['Position'] = top10.apply(lambda row: position_map.get((row['FullName'], row['round']), None), axis=1)

fig3 = px.line(
    top10,
    x='event', y='Points', color='FullName',
    markers=True,
    hover_data=['Position'],
    title="Bump chart : Evolution du classement pilotes (top 10) par Grand Prix",
    labels={'event': 'Grand Prix', 'Points': 'Points cumul√©s', 'FullName': 'Pilote', 'Position': 'Position finale'}
)
fig3.update_layout(height=700, 
    template="plotly_dark",
    font=dict(family="Montserrat, Arial", size=14),
)
fig3.show()
```

### üåç Emissions CO‚ÇÇ logistique par segment
Analyse visuelle de l'empreinte carbone F1 par trajet logistique.

```python
# Tri des segments par √©mission d√©croissante
df_flights_sorted = df_flights.sort_values('CO2_tonnes', ascending=True)

# Construction d‚Äôun label explicite pour chaque segment (√©v√®nement de->√©v√®nement √†)
df_flights_sorted['segment'] = df_flights_sorted['event_from'] + ' ‚Üí ' + df_flights_sorted['event_to']

fig = px.bar(
    df_flights_sorted,
    x='CO2_tonnes',
    y='segment',
    orientation='h',
    text='CO2_tonnes',
    color='CO2_tonnes',
    color_continuous_scale='OrRd',
    title="Emissions CO‚ÇÇ logistique par segment F1 (t¬∑CO‚ÇÇ)",
    labels={'CO2_tonnes': 't CO‚ÇÇ', 'segment': 'Trajet (GP ‚Üí GP)'}
)
fig.update_layout(height=600, xaxis_title='Tonnes de CO‚ÇÇ', yaxis_title='Segment logistique', coloraxis_showscale=False, template='plotly_dark')
fig.update_traces(texttemplate='%{x:.0f}', textposition='outside')
fig.show()
```
